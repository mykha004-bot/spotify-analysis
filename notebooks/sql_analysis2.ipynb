```python
df = pd.read_csv("../data/spotify_clean.csv")
conn = sqlite3.connect("../data/spotify.db")
df.to_sql('tracks', conn, if_exists='replace', index=False)
```

```python
pd.read_sql("""
    SELECT track_genre,
           ROUND(AVG(popularity), 1) AS avg_popularity,
           COUNT(*) AS track_count
    FROM tracks
    GROUP BY track_genre
    ORDER BY avg_popularity DESC
    LIMIT 10
""", conn)
```

```python
pd.read_sql("""
    SELECT explicit,
           ROUND(AVG(popularity), 1) AS avg_popularity,
           COUNT(*) AS track_count
    FROM tracks
    GROUP BY explicit
""", conn)
```

```python
pd.read_sql("""
    SELECT key, mode,
           ROUND(AVG(popularity), 1) AS avg_popularity,
           COUNT(*) AS track_count
    FROM tracks
    GROUP BY key, mode
    ORDER BY avg_popularity DESC
    LIMIT 10
""", conn)
```

```python
pd.read_sql("""
    SELECT track_name, artists, track_genre, popularity
    FROM tracks
    ORDER BY popularity DESC
    LIMIT 10
""", conn)
```

```python
pd.read_sql("""
    SELECT DISTINCT(track_name), artists, track_genre, popularity
    FROM tracks
    ORDER BY popularity DESC
    LIMIT 10
""", conn)
```

```python
pd.read_sql("""
    SELECT track_name, artists, track_genre, popularity
    FROM tracks
    GROUP BY track_name
    ORDER BY popularity DESC
    LIMIT 10
""", conn)
```

```python
pd.read_sql("""
    SELECT AVG(popularity) as avg_popularity
    FROM tracks
    WHERE track_genre = 'rock'
""", conn)
```
