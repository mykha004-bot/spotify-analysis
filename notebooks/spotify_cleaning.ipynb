```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
spotify_tracks = pd.read_csv("../data/spotify-tracks-dataset.csv")
spotify_tracks.head()
```

```python
spotify_tracks = spotify_tracks.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
```

```python
spotify_tracks.head()
```

```python
spotify_tracks.info()
```

```python
spotify_tracks = spotify_tracks.dropna()
spotify_tracks.info()
```

```python
spotify_tracks.head()
```

```python
spotify_tracks['duration_min'] = spotify_tracks['duration_ms'] / 60000
spotify_tracks.head()
```

```python
spotify_tracks = spotify_tracks.drop(columns=['duration_min'])
spotify_tracks.head()
```

```python
spotify_tracks['duration_min'] = (spotify_tracks['duration_ms'] / 60000).round(2)
spotify_tracks.head()
```

```python
duplicated_rows = spotify_tracks[spotify_tracks.duplicated(keep=False)]
duplicated_rows
```

```python
spotify_tracks = spotify_tracks.drop_duplicates(keep=False)
spotify_tracks.head(10)
```

```python
spotify_tracks.to_csv("../data/spotify_clean.csv", index=False)
```
