```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/spotify_clean.csv')
df.head()
```

Top 10 genres by popularity (bar chart)

```python
top_genres = df.groupby('track_genre')['popularity'].mean().nlargest(10)
top_genres.plot(kind='barh', color='#1DB954', figsize=(10, 6))
plt.title('Top 10 Genres by Average Popularity')
plt.xlabel('Average Popularity index')
plt.ylabel('Genre')
plt.tight_layout()
plt.savefig('../output/charts/top_genres.png', dpi=150)
```

```python
plt.figure(figsize=(10, 6))
sc = plt.scatter(df['energy'], df['danceability'], 
                 c=df['popularity'], cmap='YlGn', alpha=0.4, s=10)
plt.colorbar(sc, label='Popularity')
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.title('Energy vs Danceability (colored by Popularity)')
plt.savefig('../output/charts/energy_dance.png', dpi=150)
```

```python
df['popularity'].plot(kind='hist', bins=30, color='#1DB954', edgecolor='black')
plt.title('Distribution of Track Popularity Scores')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.savefig('../output/charts/popularity_dist.png', dpi=150)
```
